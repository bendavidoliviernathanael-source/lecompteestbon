# Le compte est bon
Le compte est bon est un jeu qui introduit une collection de nombres et un résultat attendu.
Ce jeu consiste à chercher une séquence de calculs qui permet d'obtenir un résultat qui est égal ou le plus proche possible du résultat attendu.
Cette séquence de calculs doit être élaborée avec des nombres de la collection de nombres et avec des opérations parmi +, -, x et /.



```python
#!/usr/bin/python3
from math import *
import doctest
import time
import tkinter as tk
from tkinter import messagebox, simpledialog
```

Dans ce programme, un calcul est une séquence constituée de trois éléments, deux nombres et un opérateur.
Des opérations pour copier une séquences de nombres, une séquence représentant un calcul et une séquence de calculs ont été définies. 

```python
def copierSequenceNombres(sequenceNombres,sequence):
 """
 Copie d'une séquence de nombres dans une séquence réceptacle

 Requis : 
 -sequenceNombres --> une séquence de nombres
 ex : [1,2,5,9]
 
 -sequence --> une séquence
 ex : []

 Fourni : 
 
 >>> copierSequenceNombres([1,2], [])
 
 """
 sequence.clear()
 cpt=0
 while(cpt<len(sequenceNombres)):
  sequence.append(sequenceNombres[cpt])
  cpt=cpt+1
```

```python
def copierCalcul(calcul,sequence):
 """
 Copie d'un calcul dans une séquence réceptacle

 Requis : 
 -calcul --> un calcul
 ex : [1,2,"+"]
 
 -sequence --> une séquence
 ex : []

 Fourni : 
 
 >>> copierCalcul([1,2,"+"], [])
 
 """
 sequence.clear
 cpt=0
 while(cpt<len(calcul)):
  sequence.append(calcul[cpt])
  cpt=cpt+1
```

```python
def copierSequenceCalculs(sequenceCalculs,sequence):
 """
 Copie d'une séquence de calculs dans une séquence réceptacle

 Requis : 
 -sequenceCalculs --> une séquence de calculs
 ex : [[1,2,"+"],[5,9,"+"]]
 
 -sequence --> une séquence
 ex : []

 Fourni : 
 
 >>> copierSequenceCalculs([[1,2,"+"]], [])
 
 """
 sequence.clear()
 cpt=0
 while(cpt<len(sequenceCalculs)):
  calcul=[]
  copierCalcul(sequenceCalculs[cpt],calcul)
  sequence.append(calcul)
  cpt=cpt+1
```

```python
def realiserCalcul(calcul):
 """
 Réalisation d'un calcul

 Requis : 
 -calcul --> un calcul
 ex : [1,2,'+']

 Fourni : 
 -un nombre
 ex : 3

 >>> realiserCalcul([1,2,'+'])
 3
 """
 res = -1
 operande1, operande2, operateur = calcul
 if(operateur=='+') : return operande1+operande2
 if(calcul[2]=='-') : return operande1-operande2
 if(calcul[2]=='*') : return operande1*operande2
 if(calcul[2]=='/') : return int(operande1/operande2)  
 return res  
```
